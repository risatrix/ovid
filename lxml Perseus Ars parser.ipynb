```python
import json
import requests

from lxml import etree
from collections import OrderedDict

data = OrderedDict()

author_dict=OrderedDict()
author_dict['author'] = "various"
author_dict['source'] = "Perseus"
author_dict['source url'] = "http://www.perseus.tufts.edu/hopper/text?doc=Perseus%3Atext%3A1999.02.0069%3Atext%3DArs"
author_dict['notes url'] = "http://www.perseus.tufts.edu/hopper/text?doc=Perseus:text:1999.02.0069:text=intro:section=2"

data.update(author_dict)

work_dict = OrderedDict()
work_dict['title'] = 'The Art of Love'

root_url = "http://www.perseus.tufts.edu/hopper/xmlchunk?doc=Perseus%3Atext%3A1999.02.0069%3Atext%3DArs%3Abook%3D"

book_list = []

# might as well define this inclusive range since we'll be using it a lot
range1 = lambda start, end: range(start, end+1)

for i in range1(1,3):
    # create structures to hold data
    bookDict = OrderedDict()
    bookDict['index'] = i
    book_str = str(i)
    bookDict['title'] = "Book " + book_str
    
    poem_list = []
    
    poemDict = OrderedDict()
    poemDict['index'] = 1
    poemDict['title'] = "Poem 1"
    
    line_list = []
    
    #use requests get xml of translation
    target_url = root_url + book_str
    print target_url
    r = requests.get(target_url)
    print r.status_code
    print r.encoding

    doc = r.content
    root = etree.fromstring(doc)

    # remove note elements
    etree.strip_elements(root, 'note', with_tail=False)
    # remove milestone elements
    etree.strip_elements(root, 'milestone', with_tail=False)
    # strip tags from entity elements but keep content
    etree.strip_tags(root,'placeName')

    lines = root.xpath('//l')
    
   
    counter = 0 
    for line in lines:
        lineDict = OrderedDict()
        counter +=1
        lineDict['index'] = counter
        lineDict['text'] = line.text
#         print line.text
        line_list.append(lineDict)
    poemDict['lines'] = line_list
    poem_list.append(poemDict)
    bookDict['poems'] = poem_list
    book_list.append(bookDict)

work_dict['books'] = book_list
data['works'] = work_dict

json_corpus = json.dumps(data)

# print json_corpus

with open('ars_translation.json', 'w') as outfile:
    json.dump(json_corpus, outfile)
    
# etree.tostring(root)
# work_dict = {'author': [{'name': 'Perseus'}, {'source':'https://en.wikisource.org/wiki/Translation:Amores'}],
#              'works': [{'title':'Ars'}, 
#                     {"books": [{"book_index": 1, "book_title": "Liber primus", 
#                                 "poems": poems_list}]}]}


#http://stackoverflow.com/questions/37046511/how-to-prevent-lxml-remove-method-from-removing-text-between-two-elements
```
